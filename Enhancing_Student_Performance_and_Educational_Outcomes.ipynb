## **Detect & Remove Outliers (Using IQR Method)**

```r
# Load required package
library(ggplot2)

# List of numeric columns to check for outliers
num_cols <- c("Age at enrollment", "Curricular units 1st sem (grade)", "Curricular units 2nd sem (grade)",
              "Unemployment rate", "Inflation rate", "GDP")

# Boxplots to visualize outliers
for (col in num_cols) {
  print(
    ggplot(dataset, aes(y = !!sym(col))) +
      geom_boxplot(fill = "steelblue", outlier.color = "red") +
      theme_minimal() +
      ggtitle(paste("Boxplot of", col))
  )
}

# Function to remove outliers using IQR
remove_outliers <- function(df, column) {
  Q1 <- quantile(df[[column]], 0.25, na.rm = TRUE)
  Q3 <- quantile(df[[column]], 0.75, na.rm = TRUE)
  IQR_value <- Q3 - Q1

  lower_bound <- Q1 - 1.5 * IQR_value
  upper_bound <- Q3 + 1.5 * IQR_value

  df <- df[df[[column]] >= lower_bound & df[[column]] <= upper_bound, ]
  return(df)
}

# Apply outlier removal to numeric columns
for (col in num_cols) {
  dataset <- remove_outliers(dataset, col)
}

# Check dataset dimensions after outlier removal
dim(dataset)
```

## **Remove Highly Correlated Features**

```r
# Install & load caret package
install.packages("caret")
library(caret)

# Compute correlation matrix for numeric variables
numeric_vars <- dataset %>% select_if(is.numeric)
cor_matrix <- cor(numeric_vars, use = "complete.obs")

# Identify & remove highly correlated features (threshold > 0.85)
high_corr <- findCorrelation(cor_matrix, cutoff = 0.85)
dataset <- dataset[ , -high_corr]  # Remove redundant features

# Check remaining features
dim(dataset)
```
