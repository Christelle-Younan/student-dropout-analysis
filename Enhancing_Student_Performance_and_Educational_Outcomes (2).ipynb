## **Detect & Remove Outliers (Using IQR Method)**

```r
# Load required package
library(ggplot2)

# List of numeric columns to check for outliers (EXCLUDING 'Nationality')
num_cols <- c("Age at enrollment", "Curricular units 1st sem (grade)", "Curricular units 2nd sem (grade)",
              "Unemployment rate", "Inflation rate", "GDP")

# Boxplots to visualize outliers
for (col in num_cols) {
  print(
    ggplot(dataset, aes(y = !!sym(col))) +
      geom_boxplot(fill = "steelblue", outlier.color = "red") +
      theme_minimal() +
      ggtitle(paste("Boxplot of", col))
  )
}

# Function to remove outliers using IQR
remove_outliers <- function(df, column) {
  Q1 <- quantile(df[[column]], 0.25, na.rm = TRUE)
  Q3 <- quantile(df[[column]], 0.75, na.rm = TRUE)
  IQR_value <- Q3 - Q1

  lower_bound <- Q1 - 1.5 * IQR_value
  upper_bound <- Q3 + 1.5 * IQR_value

  df <- df[df[[column]] >= lower_bound & df[[column]] <= upper_bound, ]
  return(df)
}

# Apply outlier removal to numeric columns
for (col in num_cols) {
  dataset <- remove_outliers(dataset, col)
}

# Check dataset dimensions after outlier removal
dim(dataset)

```

## **Remove Highly Correlated Features**

```r
# Install & load caret package
install.packages("caret")
library(caret)

# Remove columns with zero variance (constant columns)
numeric_vars <- dataset %>% select_if(is.numeric)
numeric_vars <- numeric_vars[, apply(numeric_vars, 2, var, na.rm = TRUE) > 0]

# Compute correlation matrix (handling missing values)
cor_matrix <- cor(numeric_vars, use = "pairwise.complete.obs")

# Identify & remove highly correlated features (threshold > 0.85)
if (any(is.na(cor_matrix))) {
  cor_matrix[is.na(cor_matrix)] <- 0  # Replace NA with 0 to avoid errors
}
high_corr <- findCorrelation(cor_matrix, cutoff = 0.85)

# Get column names of highly correlated variables
high_corr_vars <- colnames(numeric_vars)[high_corr]

# Ensure 'Age at enrollment' and 'Nacionality' are kept
high_corr_vars <- setdiff(high_corr_vars, c("Age at enrollment", "Nacionality"))

# Remove only the selected highly correlated variables
dataset <- dataset %>% select(-all_of(high_corr_vars))

# Check remaining features
dim(dataset)

# Verify if 'Age at enrollment' and 'Nacionality' are still there
colnames(dataset)
summary(dataset$`Age at enrollment`)
summary(dataset$`Nacionality`)


```

**Train-Test Split**

We split the data into 80% training, 20% testing for fair evaluation.

```r
# Install and load caret (if not already done)
install.packages("caret")
library(caret)

# Ensure Target is a factor
dataset$Target <- as.factor(dataset$Target)

# Create Train-Test Split (80% training, 20% testing)
set.seed(123)  # For reproducibility
train_index <- createDataPartition(dataset$Target, p = 0.8, list = FALSE)

# Split the data
train_data <- dataset[train_index, ]
test_data  <- dataset[-train_index, ]

# Check sizes
dim(train_data)
dim(test_data)
```

# 1. **Logistic Regression**

```r
# Install & load package
install.packages("nnet")
library(nnet)

# Ensure Target is a factor
train_data$Target <- as.factor(train_data$Target)
test_data$Target <- as.factor(test_data$Target)

# Train multinomial logistic regression
multi_log_model <- multinom(Target ~ ., data = train_data)

# Predict
multi_log_pred <- predict(multi_log_model, newdata = test_data)

# Evaluate
library(caret)
confusionMatrix(multi_log_pred, test_data$Target)
```
